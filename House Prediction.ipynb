# **Combine Train and Test**

```python
df = pd.concat([df_train, df_test], axis='rows')
df.shape
```

**One-hot encoding** is a common technique used to convert categorical variables into a format that can be provided as input to machine learning algorithms.

**One-hot encoding** converts categorical variables into a binary representation where each category becomes a separate binary feature column.

```python
def category_onehot_multcols(multcolumns):
    df_final=df
    i=0
    for fields in multcolumns:
        print(fields)
        df1=pd.get_dummies(df[fields],drop_first=True, dtype=int)

        df.drop([fields],axis=1,inplace=True)
        if i==0:
            df_final=df1.copy()
        else:
            df_final=pd.concat([df_final,df1],axis=1)
        i=i+1

    df_final=pd.concat([df,df_final],axis=1)

    return df_final
```

```python
categorical_columns=['MSZoning','Street','LotShape','LandContour','Utilities','LotConfig','LandSlope','Neighborhood',
         'Condition2','BldgType','Condition1','HouseStyle','SaleType',
        'SaleCondition','ExterCond',
         'ExterQual','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2',
        'RoofStyle','RoofMatl','Exterior1st','Exterior2nd','Heating','HeatingQC',
         'CentralAir',
         'Electrical','KitchenQual','Functional',
         'FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','PavedDrive']
```

```python
final_df=category_onehot_multcols(categorical_columns)
```

```python
final_df.shape
```

1. Removing duplicate columns.
2. Ensure that each column in the DataFrame final_df is unique. This is important because duplicate columns can lead to redundancy and potentially affect the performance of downstream analysis or machine learning models.

```python
final_df =final_df.loc[:,~final_df.columns.duplicated()]
```

```python
final_df.shape
```

```python
df_train= final_df.iloc[:1460, :]
df_test = final_df.iloc[1460:, :]
```

```python
df_test.drop(['SalePrice'], axis='columns',inplace=True)
```

```python
X_train = df_train.drop(['SalePrice'], axis='columns')
y_train = df_train['SalePrice']
```

# **Use Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression

# Instantiate the Linear Regression model
classifier = LinearRegression()
```

# **Training Model**

```python
# Train the model
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(df_test)
```
